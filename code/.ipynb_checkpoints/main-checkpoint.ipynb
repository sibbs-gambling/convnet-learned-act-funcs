# CNN with Learned Activitation Functions Applied to CIFAR-10 
### Data Import

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from six.moves import cPickle 
from __future__ import print_function
import argparse
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torchvision import datasets, transforms
from torch.autograd import Variable

data_dir = '../data/cifar-10-batches-py/'
file = 'data_batch'
```

```python
nBatches = 5
nImg = 10000  # per batch
img_size = [32, 32]

# Read in and Aggregate Batches
X_bin = np.zeros((nImg * nBatches, img_size[0], img_size[1], 3)).astype("uint8")
y_bin = np.zeros((nImg *  nBatches))
for jBatch in range(nBatches):
    f = open(data_dir + file + '_' + str(jBatch+1), 'rb')
    datadict = cPickle.load(f,encoding='latin1')
    f.close()
    X = datadict["data"] 
    Y = datadict['labels']
    X = X.reshape(10000, 3, 32, 32).transpose(0,2,3,1).astype("uint8")
    Y = np.array(Y)
    
    beg = jBatch*nImg
    end = (jBatch+1)*nImg
    X_bin[beg:end,:,:,:] = X
    y_bin[beg:end] = Y

nClasses = len(np.unique(Y))
```

```python
#Visualizing CIFAR 10
fig, axes1 = plt.subplots(5,15,figsize=(15,5))
for j in range(5):
    for k in range(15):
        i = np.random.choice(range(len(X)))
        axes1[j][k].set_axis_off()
        axes1[j][k].imshow(X_bin[i:i+1][0], interpolation='none')
```

### Standard CNN Implementation (Benchmark)

```python
# Training settings
parser = argparse.ArgumentParser(description='CIFAR-10 CNN')
parser.add_argument('--batch-size', type=int, default=100, metavar='N',
                    help='input batch size for training (default: 64)')
parser.add_argument('--test-batch-size', type=int, default=10000, metavar='N',
                    help='input batch size for testing (default: 1000)')
parser.add_argument('--epochs', type=int, default=20, metavar='N',
                    help='number of epochs to train (default: 10)')
parser.add_argument('--lr', type=float, default=0.01, metavar='LR',
                    help='learning rate (default: 0.01)')
parser.add_argument('--momentum', type=float, default=0.5, metavar='M',
                    help='SGD momentum (default: 0.5)')
parser.add_argument('--no-cuda', action='store_true', default=False,
                    help='disables CUDA training')
parser.add_argument('--seed', type=int, default=1, metavar='S',
                    help='random seed (default: 1)')
parser.add_argument('--log-interval', type=int, default=10, metavar='N',
                    help='how many batches to wait before logging training status')

args = parser.parse_args('')
args.cuda = not args.no_cuda and torch.cuda.is_available()

torch.manual_seed(args.seed)
if args.cuda:
    torch.cuda.manual_seed(args.seed)
```

```python
#This just defines data loaders for the training and test data.
#Datasets and data loaders turn out to be incredibly useful for setting up
#your training pipeline.  Note that you can compose a series of transformations
#of the data that will be carried out for you when data are sampled.
kwargs = {'num_workers': 4, 'pin_memory': True} if args.cuda else {}
train_loader = torch.utils.data.DataLoader(
    datasets.CIFAR10('../data', train=True, download=False,
                   transform=transforms.Compose([
                       transforms.ToTensor()
                   ])),
    batch_size=args.batch_size, shuffle=True, **kwargs)

test_loader = torch.utils.data.DataLoader(
    datasets.CIFAR10('../data', train=False, download=False, 
                     transform=transforms.Compose([
                       transforms.ToTensor()
                   ])),
    batch_size=args.test_batch_size, shuffle=True, **kwargs)
```

```python
#Here is the meat of the network definition.  Notice how easy it is!  We define
#the layers in the initialization function, and then just compose them together
#in forward.  The machinery takes care of all the tracking to work out the
#backward function for you.
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        #First layer will be a 5x5 kernel, taking in one number and outputting
        #10 (this is like saying h_1 in our slides is a length 10 vector
        self.conv1 = nn.Conv2d(3, 10, kernel_size=5)
        #Second layer is another 5x5 kernel, taking in the 10 numbers from the 
        #previous layer at that location and 
        self.conv2 = nn.Conv2d(10, 20, kernel_size=5)
        #Next we are going to do dropout on entire input channels
        self.conv2_drop = nn.Dropout2d()
        #Finally, we need some linear layers
        self.fc1 = nn.Linear(500, 50)
        self.fc2 = nn.Linear(50, nClasses)

    def forward(self, x):
        #All of these lines are just defining a sequence of compositions of the
        #functions from the initialization.  In the default example, these are
        #stacked into fewer lines.  I've unraveled them so that each can be
        #commented.

        #Compute the first layer of convolutions of the input
        x = self.conv1(x)
        #Carry out maxpooling over a 2x2 window
        x = F.max_pool2d(x, 2)
        #Take the max with 0
        x = F.relu(x)
        #Take the next convolution layer of those
        x = self.conv2(x)
        #Now do dropout on the result!
        x = self.conv2_drop(x)
        #Finally max pool these
        x = F.max_pool2d(x, 2)
        #And take the max with 0
        x = F.relu(x)
        #Before sending this output into the linear function, we reshape
        #the array of values to have 320 columns.  The -1 lets that dimension
        #be whatever is necessary to fit.
        x = x.view(-1, 500)
        #Take a linear transformation
        x = self.fc1(x)
        #Take max with 0
        x = F.relu(x)
        #Do dropout node by node.  Only do this during training. No need to
        #hide things during testing, would just give worse performance
        x = F.dropout(x, training=self.training)
        #Take a linear transofmration
        x = self.fc2(x)
        #Return the log_softmax (because we are doing multi-class
        #classification, so this will behave like log class probabilities
        return F.log_softmax(x, dim=1)

model = Net()
if args.cuda:
    model.cuda()

#Set up a stochastic gradient descent optimizer.  This takes care of the
#gradient steps for you, rather than incrementing the parameters yourself.
optimizer = optim.SGD(model.parameters(), lr=args.lr, momentum=args.momentum)

#The training is done in epochs - passes through the training data set
#Each epoch is broken up into batches which are subsets of the data set
#The train_loader produces a sequence of batches to train on.
def train(epoch):
    model.train()
    for batch_idx, (data, target) in enumerate(train_loader):
        if args.cuda:
            data, target = data.cuda(), target.cuda()
        data, target = Variable(data), Variable(target)
        #Reset all the gradients
        optimizer.zero_grad()
        #Evaluate the current model
        output = model(data)
        #Compute the loss
        loss = F.nll_loss(output, target)
        #Compute all of the needed gradients
        loss.backward()
        #Take a step in the parameters with those gradients
        optimizer.step()
        #Print out performance
        if batch_idx % args.log_interval == 0:
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
                epoch, batch_idx * len(data), len(train_loader.dataset),
                100. * batch_idx / len(train_loader), loss.data[0]))

#Evaluate performance on the test set
def test():
    model.eval()
    test_loss = 0
    correct = 0
    for data, target in test_loader:
        if args.cuda:
            data, target = data.cuda(), target.cuda()
        data, target = Variable(data, volatile=True), Variable(target)
        output = model(data)
        test_loss += F.nll_loss(output, target, size_average=False).data[0] # sum up batch loss
        pred = output.data.max(1, keepdim=True)[1] # get the index of the max log-probability
        correct += pred.eq(target.data.view_as(pred)).long().cpu().sum()

    test_loss /= len(test_loader.dataset)
    print('\nTest set: Average loss: {:.4f}, Accuracy: {}/{} ({:.0f}%)\n'.format(
        test_loss, correct, len(test_loader.dataset),
        100. * correct / len(test_loader.dataset)))
```

```python
#Carry out training for desired number of epochs
for epoch in range(1, args.epochs + 1):
    train(epoch)
    test()
```
